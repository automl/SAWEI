```python
from __future__ import annotations
from pathlib import Path

content = r"""
\begin{figure}[h]
    \centering
    \includegraphics[width=0.85\linewidth]{filename}
    \caption{captionstr}
    \label{labelstr}
\end{figure}
"""
benchmark = "BBOB"
paths = list(Path(f"figures/{benchmark}").glob("alpha/*.pdf"))
paths.sort()

C = []
for p in paths:
    if benchmark == "BBOB":
        caption = f"BBOB Function {int(p.stem)}" # + {int(p.stem) + 1}"
    elif benchmark.startswith("HPOBench"):
        caption = f"HPOBench ML: (model, task\_id) = {p.stem}"
    else:
        raise ValueError
    replacements = {
        "labelstr": f"fig:{p}",
        "filename": str(p),
        "captionstr": caption,
    }
    new_content = content
    for k, v in replacements.items():
        new_content = new_content.replace(k, v)
        
    C.append(new_content)
    
C = "\n".join(C)

print(C)
    

```
