```python
from __future__ import annotations
from pathlib import Path

content = r"""
\begin{figure}[h]
    \centering
    \includegraphics[width=\linewidth]{filename}
    \caption{captionstr}
    \label{labelstr}
\end{figure}
"""

paths = list(Path("figures").glob("log_regret/*.png"))
paths.sort()

C = []
for p in paths:
    caption = f"BBOB Function {int(p.stem)} + {int(p.stem) + 1}"
    replacements = {
        "labelstr": f"fig:{p}",
        "filename": str(p),
        "captionstr": caption,
    }
    new_content = content
    for k, v in replacements.items():
        new_content = new_content.replace(k, v)
        
    C.append(new_content)
    
C = "\n".join(C)

print(C)
    

```
